```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df.head()
```

```python
# top 5 zipcodes for 911 calls 
df['zip'].value_counts().head(5)
```

```python
#top 5 townships (twp) for 911 calls
df['twp'].value_counts().head(5)
```

```python
#how many unique title codes
df['title'].nunique()
```

```python
# create Reason Column
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
#Most Common Reson
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=df)
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
sns.countplot(x='Day of Week',data=df,hue='Reason')
```

```python
sns.countplot(x='Month',data=df,hue='Reason')
# missing months 9,10 and 11
```

```python
byMonth = df.groupby('Month').count()
byMonth
```

```python
#try simple line plot that fills in the missing months
byMonth['twp'].plot()
```

```python
sns.lmplot(x='Month',y='twp',data=byMonth.reset_index())
```

```python
df['Date']=df['timeStamp'].apply(lambda t:t.date())
```

```python
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.title('EMS')
```

```python
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.title('Traffic')
```

```python
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.title('Fire')
```

```python
dayHour = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='Reds')
```

```python
sns.clustermap(dayHour,cmap='Reds')
```

```python
dayMonth = df.groupby(by=['Day of Week','Month']).count()['desc'].unstack()
dayMonth.head()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(dayMonth,cmap='viridis')
plt.tight_layout()
```

```python
sns.clustermap(dayMonth,cmap='viridis')
```
